```python
#indexing 索引 
import torch 
a=torch.rand(2,3,4,5)
print(a)
print(a[0].shape)
print(a[0,0].shape)
print(a[0,0,3,4])
```

```python
# ：的应用
import torch
a=torch.rand(3,3,4,5) #3张照片 rgb通道 每个通道4*5
#print(a)
print(a.shape) #打印a的维度 ，四个数，四个维度，每一维度下的数据
print(a[:2].shape) #0->2 [0,2) 将第1张和第二张照片的格式打印出来
print(a[:2,:1,:,:].shape) #第一 第二张照片的 r通道 每个像素点 

# ：再数的后面
print(a[:2,1:,:,:].shape) #第一和第二张照片 的g、b通道每个像素点
print(a[:2,-3:,:,:].shape) #第一和第二张照片，rgb三个通道的每个像素点
```

```python
#：隔行采样
import torch
a=torch.rand(3,3,4,5)
print(a[:,:,0:4:2,0:5:2].shape)#取不到最后一个元素
```

```python
import torch
a=torch.rand(3,3,4,5)
print(a[:,:,::2,::2].shape)
```

```python
#再特定的维度上的索引进行操作
import torch
a=torch.rand(3,3,4,5)
print(a)
#a.shape
print(a.index_select(0,torch.tensor([0,2])).shape) #对0索引（照片）上第1张和第3张照片 像素点输出
print(a.index_select(0,torch.tensor([0,2])))
```

```python
#再特定的维度上的索引进行操作 rgb通道上的操作
import torch
a=torch.rand(3,3,4,5)
print(a)
#a.shape
print(a.index_select(1,torch.tensor([0,1])).shape) #对1索引（通道）中的 RG通道的像素点进行输出 
print(a.index_select(1,torch.tensor([0,1])))
```

```python
#再特定的维度上的索引进行操作 照片高度上的操作
import torch
a=torch.rand(3,3,4,5)
print(a)
#a.shape
print(a.index_select(2,torch.arange(4)).shape)#对2索引（照片）中的h进行操作 输出【0,1,2,3】
print(a.index_select(2,torch.arange(4)))
```

```python
#...省略号的使用
import torch
a=torch.rand(3,3,4,5)
a[...].shape
```

```python
print(a[...,:2]) #输出3张照片每个rgb通道下的前两列
a[...,:2].shape
```

```python
#select by musk 选出tensor中某个大于或者小于某个阈值的值
import torch
x=torch.rand(4,4)
print(x)

mask=x.ge(0.5)
print(mask)

Y=torch.masked_select(x,mask)
print(Y)

Y.shape
```
