```python
# cat  拼接 拼接的tensor有一个维度不同，其余维度相同 维度相加
import torch
a=torch.rand(4,32,8) #四个班级，一个班级32名学生.一名学生8门课成绩
#print(a)
b=torch.rand(5,32,8)
torch.cat([a,b],dim=0).shape #0维度上（第一个维度）拼接，九个班级.一个班级32名学生，一名学生8门课成绩
```

```python
#example 2张照片 三个通道  扩展成4个通道
import torch
a=torch.rand(2,3,4,5)
b=torch.rand(2,1,4,5)
print(a)
print(b)
c=torch.cat([a,b],dim=1)
print(c)
```

```python
#stack 维度保持一致，维度扩充，增加维度
import torch
a=torch.rand(8,5)
b=torch.rand(8,5)
c=torch.stack([a,b],dim=0)
print(c)
torch.stack([a,b],dim=0).shape #一个老师教的两个班的学生成绩表，合并到一个成绩表中
```

```python
#split 按长度拆分
import torch
a=torch.rand(5,8)
b=torch.rand(5,8)
c=torch.stack([a,b],dim=0)
print(c.shape)
a1,b1=c.split(1,dim=0) #2/1=2  按每个长度是一，可以拆分成两个
print(a1,"\n",b1)
```

```python
#split 按长度拆分 example
import torch
a=torch.rand(3,8,5)
b=torch.rand(2,8,5)
c=torch.cat([a,b],dim=0)
print(c.shape)
a1,b1=c.split([3,2],dim=0) #第一个三个长度，第二个两个长度
print(a1)
print(b1)
```

```python
#chunk 按num拆分
import torch
a=torch.rand(5,8)
b=torch.rand(5,8)
c=torch.stack([a,b],dim=0)
print(c.shape)
c1,d1=c.chunk(2,dim=0) #拆分两个
print(c1)
print(d1)
```
